## 3. Control flow and functions
A tiny credit rule as a function — the kind of logic a *rule-based* scorecard uses before we ever reach machine learning.

```python
def simple_decision(dti, utilization):
    """Toy rule: reject if debt-to-income OR utilization is too high."""
    if dti > 0.5 or utilization > 0.9:
        return "reject"
    return "accept"

for dti in [0.3, 0.6]:
    print(f"dti={dti} -> {simple_decision(dti, 0.4)}")
```

## 4. NumPy in 60 seconds
NumPy arrays let us do maths on whole columns at once (vectorisation) instead of looping. pandas is built on top of NumPy.

```python
import numpy as np
balances = np.array([1200, 5400, 300, 9800])
print("mean:", balances.mean(), "| max:", balances.max())
print("with 2% interest:", balances * 1.02)
```

### 5b. Selecting columns and rows (filtering)

```python
# one column (a Series)
print(df["loan_amount_pln"].describe())

# rows that satisfy a condition (boolean mask)
high_dti = df[df["debt_to_income"] > 0.5]
print("\nHigh-DTI customers:", len(high_dti))
```

### 5c. Group-by: the single most useful pandas verb
*'What is the average default rate by housing status?'* is a group-by.

```python
df.groupby("housing")["default"].mean().round(3)
```
